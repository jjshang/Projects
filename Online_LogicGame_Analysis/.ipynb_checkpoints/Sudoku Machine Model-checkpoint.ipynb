```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import linear_model
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
import datetime as dt
```

```python
sudoku = pd.read_csv("sudoku_counts.csv")
```

```python
sudoku["Age"] = int(dt.datetime.today().strftime("%Y")) - sudoku["Birth_Year"]
```

```python
sudoku["Gender_numeric"] = sudoku["Gender"].map({"m":1, "f":0})
```

```python
sudoku.head()
```

```python
sudoku = sudoku.drop(sudoku[sudoku.Age > 90].index)
sudoku = sudoku.drop(sudoku[sudoku.Age < 6].index)
sudoku = sudoku.dropna()
```

```python
print(sudoku.Gender.unique())
print(sudoku.Age.sort_values().unique())
```

```python
sudoku.head()
```

```python
sudoku.shape
```

```python
gender_dummies = pd.get_dummies(sudoku.Gender)
```

```python
sudoku_best_time = pd.DataFrame(sudoku["Best Time (s)"])
sudoku_best_time = sudoku_best_time.join(sudoku.Age).join(gender_dummies)
sudoku_best_time.head()
```

```python
Y = sudoku["Best Time (s)"]
X = sudoku_best_time.drop("Best Time (s)", axis = 1)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = .20, random_state = 40)
regr = linear_model.LinearRegression() # Do not use fit_intercept = False if you have removed 1 column after dummy encoding

regr.fit(X_train, Y_train)
predicted = regr.predict(X_test)
predicted
```

```python
X1 = sm.add_constant(X_train)
model = sm.OLS(Y_train, X1).fit() 
```

```python
model.summary()
```

## Can we get model that predicts number of puzzles solved based on Age, Gender

```python
## Level 1

sudoku1 = sudoku[sudoku["Puzzle Level"] == 1]
```

```python
sudoku1.head()
```

```python
### Create dummy variables for Gender and Country
gender_dummies = pd.get_dummies(sudoku1.Gender)
country_dummies = pd.get_dummies(sudoku1.Country)
```

```python
### Create the data table that has the y-value and the x-variables
sudoku1_numpuzzle = pd.DataFrame(sudoku1["Puzzles Solved"])
sudoku1_numpuzzle = sudoku1_numpuzzle.join(sudoku1.Age).join(gender_dummies)
sudoku1_numpuzzle.head()
```

```python
### Create linear regression model
Y = sudoku1["Puzzles Solved"]
X = sudoku1_numpuzzle.drop("Puzzles Solved", axis = 1)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = .20, random_state = 40)
regr = linear_model.LinearRegression() # Do not use fit_intercept = False if you have removed 1 column after dummy encoding

regr.fit(X_train, Y_train)
predicted = regr.predict(X_test)
predicted
```

```python
X1 = sm.add_constant(X_train)
model = sm.OLS(Y_train, X1).fit() 
```

```python
model.summary()
```
