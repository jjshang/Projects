```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import linear_model
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
import datetime as dt
```

```python
sales = pd.read_csv("sales-anonymized.csv")
```

```python
sales.head()
```

```python
sales["DateTime Created"] = pd.to_datetime(sales["DateTime Created"])
```

```python
sales.shape
```

```python
sales = sales.dropna(axis=0, subset=["Gender(M/F)"])
```

```python
sales["Gender_num"] = sales["Gender(M/F)"].map({"F":0, "M":1})
```

```python
sales.head()
```

```python
def date_extract(start, end):
    sub_sales = sales[(sales["DateTime Created"] > start) & (sales["DateTime Created"] < end)]
    return sub_sales
```

```python
sales1 = date_extract("2010-08", "2011-02")
```

```python
sales1.groupby("Gender(M/F)")["Gender_num"].sum()/sales1.count()[0]
```

```python
sales1.count()[0]
```

```python
sales1.groupby("Education").count()
```

```python
sales1.groupby(["Education","Status"])["Status"].count()
```
